```python
#NOTICE
# This code is adapted from Alex Fredberg


#Necessary Imports:

from bs4 import BeautifulSoup
#Allows functionality with html documents

import requests
import smtplib
import time
import datetime
```

```python
#Access Amazon

#URL of the webpage we are looking to scrape
url = 'https://www.amazon.com/AmazonBasics-Lightning-USB-Cable-Certified/dp/B082T6K8Y2/ref=sr_1_7_ffob_sspa?crid=2R73NQGIGQG4Y&keywords=iphone+charger&qid=1689019950&sprefix=iphone+charge%2Caps%2C154&sr=8-7-spons&sp_csd=d2lkZ2V0TmFtZT1zcF9tdGY&psc=1'

#Data of profile viewing the website, needed for the bf4 functionality
#Can be found by using the following website: 'http://httpbin.org/get'
headers = { "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1" }

```

```python
#Access the webpage
webpage = requests.get(url, headers= headers)

page_content = BeautifulSoup(webpage.content, 'html.parser')

page_content_formatted = BeautifulSoup(page_content.prettify(), 'html.parser')

print(page_content_formatted)
```

```python
#What are we parsing
try:
    title = page_content_formatted.find(id='productTitle').get_text()
    print(title)
except:
    print("Title Not Found")
    print("Check bot-detection")
```
